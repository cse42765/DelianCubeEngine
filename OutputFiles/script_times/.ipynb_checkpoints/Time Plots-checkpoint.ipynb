<h4>Executed the same query with data of :

    - Lines of data : 1.000
    - Lines of data : 10.000
    - Lines of data : 100.000
    - Lines of data : 1.000.000
    - Lines of data : 10.000.000
    
Using :

    - RDBMS
    - Spark with 1-10 nodes as executors
    
And executed each combination of the above for :

    - 10 times

The tests run on the cluster of the university :

    - 11 nodes : 1 master and 10 slaves
    - Each node is a SUN Fire X4100 Servers with 4 CPUs (Dual Core AMD Opteron στα 2193 MHz each CPU)
    - Each node has 12GB or RAM
    - For our tests we used 3GB on each node.
